### Linear Model: Fit a regression model (with an out-ofsample R42 of >0.01) to a financial time series. Financial series have a low signal-to-noise ratio, so even a weak correlation will require some basic feature engineering (moving average prices, volatility, etc).



```python
# import libraries 
import numpy as np
import pandas as pd
import quandl
import datetime
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
plt.style.use('seaborn-darkgrid')
plt.rc('figure', figsize=(16,10))
plt.rc('lines', markersize=4)
start_date = datetime.date(2009, 3,8)
end_date = datetime.date.today()
```

```python
# Load data from Quandl
data = quandl.get('FSE/SAP_X', start_date=start_date, end_date=end_date)

data.to_csv('data/sap_stock.csv')
data.head()
data.info()
data.describe()
#use closing to predict
df = pd.DataFrame(data, columns=['Close'])
df = df.reset_index()

df.isna().values.any()
```

```python
years = mdates.YearLocator() # Get years
yearsFmt = mdates.DateFormatter('%Y') # year format
```

```python
# Create subplots to plot graph and control axes
fig, ax = plt.subplots()
ax.plot(df['Date'], df['Close'])

# Format the ticks
ax.xaxis.set_major_locator(years)
ax.xaxis.set_major_formatter(yearsFmt)

# Set figure title
plt.title('Close Stock Price History [2009 - 2020]', fontsize=16)
# Set x label
plt.xlabel('Date', fontsize=14)
# Set y label
plt.ylabel('Closing Stock Price in $', fontsize=14)

# Rotate and align the x labels
fig.autofmt_xdate()

# Show plot
plt.show()
```

```python
# split data to train and test 
train, test = train_test_split(df, test_size=0.20)

X_train = np.array(train.index).reshape(-1, 1)
y_train = train['Close']
```

```python
# Create LinearRegression Object
model = LinearRegression()
# Fit linear model using the train data set
model.fit(X_train, y_train)
print('Slope(m): ', np.asscalar(np.squeeze(model.coef_)))
# The Intercept
print('Intercept(y): ', model.intercept_)
```

```python
plt.figure(1, figsize=(16,10))
plt.title('Linear Regression | Price vs Time')
plt.scatter(X_train, y_train, edgecolor='w', label='Actual Price')
plt.plot(X_train, model.predict(X_train), color='g', label='Predicted Price')
plt.xlabel('Integer Date')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
X_test = np.array(test.index).reshape(-1, 1)
y_test = test['Close']
y_pred = model.predict(X_test)
# Generate 25 random numbers
randints = np.random.randint(2550, size=25)

# Select row numbers == random numbers
df_sample = df[df.index.isin(randints)]

df_sample.head()

```

```python
plt.figure(1, figsize=(16,10))
plt.title('Linear Regression | Price vs Time')
plt.plot(X_test, model.predict(X_test), color='r', label='Predicted Price')
plt.scatter(X_test, y_test, edgecolor='w', label='Actual Price')

plt.xlabel('Integer Date')
plt.ylabel('Stock Price in $')

plt.show()
```

```python
df['Prediction'] = model.predict(np.array(df.index).reshape(-1, 1))
df.head()
df['Close'].describe()
```

```python
# cost functions
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
# r2
print('R2: ', metrics.r2_score(y_test, y_pred))
```

#### What are some of the weakness of the model and how can it be improved with additional work?

The value of 2 shows that are model accounts for nearly 94% of the differences between the actual stock prices and the predicted prices.However,Linear regression depends on certain underlying assumptions. Violation of these assumptions outputs poor results.
For example,It is assumed that there is little or no multicollinearity in the data.This results in unstable parameter estimates which makes it very difficult to assess the effect of independent variables.
Due to the Central Limit Theorem, we may assume that there are lots of underlying facts affecting the process and the sum of these individual errors will tend to behave like in a zero mean normal distribution.  
A single outlier can significantly change the model parameters.The model can be improved by using techniques such as regularization and model validation.We can also use polynomial regression to improve on the accuracy.
